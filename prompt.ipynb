### Required Assignment 5.1: Will the Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaurant near where you are driving. Would you accept that coupon and take a short detour to the restaurant? Would you accept the coupon but use it on a subsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaurant? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\$20 - $50).

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons. **Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
# Create a new DataFrame containing only rows with 'Bar' coupons
bar_data = data[data['coupon'] == 'Bar']

# Display the shape and first few rows
print("Shape of bar_data:", bar_data.shape)
print("\nFirst 5 rows of bar_data:")
bar_data.head()
```

2. What proportion of bar coupons were accepted?


```python
# Calculate the proportion of accepted bar coupons (Y == 1)
bar_acceptance_rate = bar_data['Y'].value_counts(normalize=True)

# Extract the proportion for Y = 1 (accepted)
bar_acceptance_proportion = bar_acceptance_rate.get(1, 0)

bar_acceptance_proportion
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
# Define the two groups based on bar visit frequency
low_freq = bar_data[bar_data['Bar'].isin(['never', 'less1', '1~3'])]
high_freq = bar_data[bar_data['Bar'].isin(['4~8', 'gt8'])]

# Calculate acceptance rate for each group
low_freq_rate = low_freq['Y'].mean()
high_freq_rate = high_freq['Y'].mean()

low_freq_rate, high_freq_rate
```

```python
# Acceptance rate of those who went 3 or fewer times was 37% while those who went to bar 4 times or more was ~77%
# This shows that for a bar coupon the chance of acceptance of coupon among those who frequent the bar more (4 or more) is 2x higher. This is expected
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
# Define conditions
bar_more_than_once = bar_data['Bar'].isin(['1~3', '4~8', 'gt8'])
age_over_25 = bar_data['age'].isin(['26', '31', '36', '41', '46', '50plus'])

# Group: go to a bar > once/month AND age > 25
target_group = bar_data[bar_more_than_once & age_over_25]

# Group: all others
other_group = bar_data[~(bar_more_than_once & age_over_25)]

# Acceptance rates
target_acceptance_rate = target_group['Y'].mean()
other_acceptance_rate = other_group['Y'].mean()

target_acceptance_rate, other_acceptance_rate
```

```python
#Drivers who go to bars more than once/month and are over age 25:
# ~69.5% acceptance rate

# All other drivers:
# ~33.5% acceptance rate
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry.


```python
# Define filtering conditions
bar_more_than_once = bar_data['Bar'].isin(['1~3', '4~8', 'gt8'])
not_with_kids = bar_data['passanger'] != 'Kid(s)'
not_farming_fishing = ~bar_data['occupation'].isin(['Farming & Fishing', 'Farming & Fishing & Forestry', 'Forestry'])

# Group: goes to bar > once/month, no kids as passenger, and not in specified occupations
target_group = bar_data[bar_more_than_once & not_with_kids & not_farming_fishing]

# Group: all others
other_group = bar_data[~(bar_more_than_once & not_with_kids & not_farming_fishing)]

# Calculate acceptance rates
target_acceptance_rate = target_group['Y'].mean()
other_acceptance_rate = other_group['Y'].mean()

target_acceptance_rate, other_acceptance_rate
```

```python

# Drivers who go to bars more than once/month, had no kids as passengers, and do not work in farming/fishing/forestry:
# ~71.3% acceptance rate

# All other drivers:
# ~29.6% acceptance rate

```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K.



```python
# Define common conditions
bar_more_than_once = bar_data['Bar'].isin(['1~3', '4~8', 'gt8'])

# First condition: bar > once/month, no kids, not widowed
cond1 = bar_more_than_once & (bar_data['passanger'] != 'Kid(s)') & (bar_data['maritalStatus'] != 'Widowed')

# Second condition: bar > once/month and age < 30
under_30 = bar_data['age'].isin(['below21', '21', '26'])
cond2 = bar_more_than_once & under_30

# Third condition: cheap restaurants > 4 times/month and income < $50K
cheap_rest_often = bar_data['RestaurantLessThan20'].isin(['4~8', 'gt8'])
low_income = bar_data['income'].isin(['Less than $12500', '$12500 - $24999', '$25000 - $37499', '$37500 - $49999'])
cond3 = cheap_rest_often & low_income

# Combine all three conditions using OR
target_group = bar_data[cond1 | cond2 | cond3]

# All other drivers
other_group = bar_data[~(cond1 | cond2 | cond3)]

# Calculate acceptance rates
target_acceptance_rate = target_group['Y'].mean()
other_acceptance_rate = other_group['Y'].mean()

target_acceptance_rate, other_acceptance_rate
```

```python
# Target Group -Drivers who met any of the following:
# Go to bars > once/month, no kids, not widowed
# Go to bars > once/month and are under 30
# Go to cheap restaurants > 4x/month and earn <$50K)
# ~58.9% acceptance rate

# All Other Drivers
# ~29.8% acceptance rate

# Drivers in the target group are nearly twice as likely to accept a bar coupon. 
# This indicates that social habits, youth, and budget-conscious behavior all increase the likelihood of accepting such coupons.
```

7.  Based on these observations, what do you hypothesize about drivers who accepted the bar coupons?

```python
# Past Behavior:
# Drivers who go to bars more than once per month are significantly more likely to accept a bar coupon.
# Suggests that past behavior is a strong predictor of future interest.
# Regular bar-goers may be more responsive to promotions in that category.

# Passenger Context
# Drivers traveling without children are much more likely to accept bar coupons.

# Age and Lifestyle Influence Behavior
# Drivers who are under 30 or not widowed show higher acceptance.

# Drivers who are socially active, younger, and traveling without children are substantially more likely to accept bar coupons.
```

### Independent Investigation

Using the bar coupon example as motivation, you are to explore one of the other coupon groups and try to determine the characteristics of passengers who accept the coupons.  

```python
# Filter data for restaurant coupons (under $20 category)
restaurant_data = data[data['coupon'] == 'Restaurant(<20)']

# Group by income and calculate the acceptance rate
restaurant_income_acceptance = restaurant_data.groupby('income')['Y'].mean().sort_index()

restaurant_income_acceptance
```

```python
# Acceptance is high across most income levels, especially for those earning $50K–$62K (~78.2%).
# Even the lowest and highest income groups show strong interest (~72%).
# A notable drop appears in the $87.5K–$99.9K range (57.8%).
```

```python
# Filter data for restaurant coupons ($20–$50 category)
restaurant_20_50_data = data[data['coupon'] == 'Restaurant(20-50)']

# Group by income and calculate the acceptance rate
restaurant_20_50_income_acceptance = restaurant_20_50_data.groupby('income')['Y'].mean().sort_index()

restaurant_20_50_income_acceptance
```

```python
# Acceptance rates are much lower overall compared to cheaper restaurant coupons.
# The highest acceptance is from the $25K–$37.5K group (~51.1%).
# The lowest income groups (under $25K) show low acceptance (36–38%), likely due to the higher price point.
```

```python
# Define income order for consistent plotting
income_order = [
    'Less than $12500', '$12500 - $24999', '$25000 - $37499',
    '$37500 - $49999', '$50000 - $62499', '$62500 - $74999',
    '$75000 - $87499', '$87500 - $99999', '$100000 or More'
]

# Acceptance rates
cheap_rest_acceptance = restaurant_income_acceptance.reindex(income_order)
expensive_rest_acceptance = restaurant_20_50_income_acceptance.reindex(income_order)

# Prepare data for plotting
plot_df = pd.DataFrame({
    'Income': income_order,
    'Restaurant (<$20)': cheap_rest_acceptance.values,
    'Restaurant ($20-$50)': expensive_rest_acceptance.values
})

# Melt to long format for seaborn
plot_df_melted = plot_df.melt(id_vars='Income', var_name='Coupon Type', value_name='Acceptance Rate')

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=plot_df_melted, x='Income', y='Acceptance Rate', hue='Coupon Type', marker='o')
plt.title('Acceptance Rates by Income: Cheap vs. Expensive Restaurant Coupons')
plt.xlabel('Income Level')
plt.ylabel('Acceptance Rate')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Acceptance rates for cheap restaurant coupons are consistently higher across all income brackets.
# Even high-income individuals are more likely to accept cheaper coupons.
```

```python
# The acceptance rate for expensive restaurant coupons is:
# Significantly lower in lower-income groups (<$50K).
# Not significantly higher in upper-income groups either.
```
